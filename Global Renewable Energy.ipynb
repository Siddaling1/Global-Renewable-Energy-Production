#  Question/ Queries

# 1.What is the total renewable energy produced by each country? 

```python
Total_Renewable = final.groupby('Country_Name')['Energy_Produced_GWh'].sum()
Total_Renewable
```

```python
Total_Renewable.plot.bar()
```

# 2.Which energy source has the highest total energy production?

```python
Total_energy_production = final.groupby('Source_Type')['Energy_Produced_GWh'].sum()
Total_energy_production
```

```python
Total_energy_production.plot.pie(colors=["red",'yellow','green','orange'])
```

# 3.What is the average CO emission for each continent?

```python
Average_co2_consumption=final.groupby("Continent")["CO2_Emissions_MTons"].mean()
Average_co2_consumption
```

```python
Average_co2_consumption.plot.line(color="red")
```

# 4.What is the trend of total global energy production by year?

```python
total_energy=final.groupby('Year')['Energy_Produced_GWh'].sum()
total_energy.sort_values(ascending=False)
```

```python
total_energy.plot.line(color='green')
```

# 5.Which continent has the highest average efficiency rate?

```python
efficiency_rate=final.groupby('Continent')['Efficiency_Rate'].mean()
efficiency_rate.sort_values(ascending=False)
```

# 6.Compare total Energy Produced vs Energy Consumed for each country.

```python
compare_energy=final.groupby('Country_Name')[['Energy_Produced_GWh','Energy_Consumed_GWh']].sum()
compare_energy
```

```python
compare_energy.plot.bar()
```

# 7.What are the top 5 countries with the lowest CO emissions?

```python
lowest_co2=final.groupby('Country_Name')['CO2_Emissions_MTons'].sum()
lowest_co2.head(5).sort_values(ascending=True)
```

```python
lowest_co2.head(5).sort_values(ascending=True).plot.barh(color='green')
```

# 8.What is the most commonly used energy source across all countries?

```python
most_used_energy = (
    final.groupby(['Country_Name', 'Source_Type'])['Energy_Produced_GWh']
    .sum()
    .reset_index()
)

most_used_energy = most_used_energy.loc[
    most_used_energy.groupby('Country_Name')['Energy_Produced_GWh'].idxmax()
].reset_index(drop=True)

most_used_energy = most_used_energy.sort_values(by='Energy_Produced_GWh', ascending=False)

print(most_used_energy)
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2580",
    database="Global_Renewable_Energy_Production"
)
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS Global_Renewable_Energy_Production (
    Record_ID_x VARCHAR(50),
    Country_ID  VARCHAR(50),
    Source_ID   VARCHAR(50),
    Year        INT(50),
    Energy_Produced_GWh FLOAT(50),
    Country_Name    VARCHAR(50),
    Continent   VARCHAR(50),
    GDP_Per_Capita  FLOAT(50),
    Population  INT(50),             
    Record_ID_y    VARCHAR(50),
    Energy_Consumed_GWh FLOAT(50),
    CO2_Emissions_MTons FLOAT(50),
    Source_Type VARCHAR(50),
    Category    VARCHAR(50),
    Efficiency_Rate FLOAT(50)
)
""")
```
